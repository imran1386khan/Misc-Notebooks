```
import numpy as np
import pandas as pd
from random import random
```

```
sample_data = [[0,0],
              [0,1],
              [1,0],
              [1,1]]

expected_results = [0,
                   1,
                   1,
                   1]
```

```
weights = np.random.random(2)/1000
weights
```

```
bias_weight = np.random.random()/1000
bias_weight
```

```
activation_threshold = 0.5
for iter_num in range(5):
  correct_answer = 0
  for idx, sample in enumerate(sample_data):
    input_vector = np.array(sample)
    weights = np.array(weights)
    activation_level = np.dot(input_vector, weights) + (bias_weights *1)
    if activation_level > activation_threshold:
      perceptron_output = 1
    else:
      perceptron_output = 0
    
    if perceptron_output == expected_results[idx]:
      correct_answer += 1
    new_weights = []
    print('Expected results: {}, perceptron_output :{}'.format(expected_results[idx], perceptron_output))
    for i, x in enumerate(sample):
      new_weights = weights + ((expected_results[idx] - perceptron_output) * x)
      print('old weight : {}, new weight : {} for sample {}'.format(weights, new_weights, sample))
    print('******' * 10)
    bias_weight = bias_weight + ((expected_results[idx] - perceptron_output) * 1)
    weights = np.array(new_weights)
    
  print('{} correct answers out of 4, for iteration {}'.format(correct_answer, iter_num))
  print('******' * 20)  
  
```

XOR Keras Implementation

```
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD
```

```
X_train = np.array([[0,0],
                   [0,1],
                   [1,0],
                   [1,1]])
y_train = np.array([[0],
                   [1],
                   [1],
                   [0]])
```

```
model = Sequential()
num_neurons = 10
model.add(Dense(num_neurons, input_dim=2))
model.add(Activation('tanh'))
model.add(Dense(1))
model.add(Activation('sigmoid'))
model.summary()
```

```
sgd = SGD(lr=0.1)
model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```
# This is just doing metrix multiplication with the inputs to the random initialized weights with no knowledge of the output that is required.
model.predict(X_train)
```

```
model.fit(X_train,y_train, epochs=100)
```

```
# Saved Trained Model
import h5py
model_structure = model.to_json()

with open("basic_model.json", "w") as json_file:
  json_file.write(model_structure)
  
model.save_weights("basic_weights.h5")
```
